```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
with open('text_articles/test_algo.txt', encoding='utf-8') as f:
    txt = ''.join(f.readline())
```

```python
wc = WordCloud()
```

```python
wc.generate(txt)
```

```python
wc.to_file('output.png')
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
